```python

import numpy as np
x=[]
y=[]
z=[]
for i in range(1001):
  x= np.random.normal(2.5,0.5,i)
  y = np.random.normal(3,0.6,i)
  z=np.random.normal(20,4,i)

import pandas as pd
dfx = pd.DataFrame(x)
dfx.to_excel(excel_writer = "/content/drive/MyDrive/1/1000X.xlsx")
dfy = pd.DataFrame(y)
dfy.to_excel(excel_writer = "/content/drive/MyDrive/1/1000Y.xlsx")
dfz = pd.DataFrame(z)
dfz.to_excel(excel_writer = "/content/drive/MyDrive/1/1000Z.xlsx")

import matplotlib.pyplot as plt
import pylab
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(x,y,color='orange')
plt.title('1000 Scenarios / X-Y')
plt. savefig('/content/drive/MyDrive/1/1000XY.png')
plt.show()

plt.xlabel("X")
plt.ylabel("Z")
plt.scatter(x,z,color='purple')
plt.title('1000 Scenarios / X-Z')
plt. savefig('/content/drive/MyDrive/1/1000XZ.png')
plt.show()

plt.xlabel("Y")
plt.ylabel("Z")
plt.scatter(y,z,color='green')
plt.title('1000 Scenarios / Y-Z')
plt. savefig('/content/drive/MyDrive/1/1000YZ.png')
plt.show()

ax = plt.axes(projection='3d')
# Data for three-dimensional scattered points
zdata = z
xdata = x
ydata = y
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Reds')
plt. savefig('/content/drive/MyDrive/1/1000XYZ.png')

# path='/content/drive/MyDrive/1/3000Scenarios.txt'

# with open(path,'w') as table:
#     for row in zip(x,y,z):
#         for cell in row:
#             table.write(str(cell) + '\t')
#         table.write('\n')
```
